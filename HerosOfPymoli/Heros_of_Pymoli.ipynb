```python
#import Dependices
import pandas as pd
#load and read file
file= "../Resources/purchase_data.csv"
df=pd.read_csv(file)
```

```python
#PLAYER COUNT
#displaying the total number of players by SN
Total_Players=(len(df["SN"].unique()))

#create a data frame with a total players column
Player_count=pd.DataFrame({"Total Players": [Total_Players]})
Player_count
```

```python
#PURCHASING ANALYSIS-TOTAL
#Basic calulation for unique items
Number_of_items=len(df["Item Name"].value_counts())

#Basic calulation for Average Price
Average_price= df["Price"].mean()
#Basic calulation for number of purchases
Number_of_purchases=df["Purchase ID"].count()
#Basic calulation for Total Revenue
Total_Revenue=df["Price"].sum()

#Create a new data frame to hold the results
Purchasing_Analysis_Total= pd.DataFrame({"Number of Unique Items": [Number_of_items], "Average_Price": [Average_price], "Number of Purchases" : [Number_of_purchases], "Total_Revenue": [Total_Revenue]})
#Format the data frame to look cleaner - converting the Average Price and Total Revenue to dollars, rounding, and adding a comma in the thousands place
Purchasing_Analysis_Total["Average_Price"] = Purchasing_Analysis_Total["Average_Price"].map("${:.2f}".format)
Purchasing_Analysis_Total["Total_Revenue"] = Purchasing_Analysis_Total["Total_Revenue"].map("${:,}".format)

#Print the Data Frame
Purchasing_Analysis_Total
```

```python
#GENDER DEMOGRAPHIS

#Counts of Genders
##Create a table with only SN and Gender to simplify; 780 show up but we know there are a total of 576 players so there must be duplicates
Genders=df[["SN", "Gender"]]
##Drop the duplicates to get total to 576
Genders=Genders.drop_duplicates()
##Count the values for the individal Genders
Genders_Count=Genders["Gender"].value_counts()

#Percentage of Players
##create list for calculation of percentages
Totals=[Genders_Count[0], Genders_Count[1], Genders_Count[2]]
##Percentage Calcluation with a rounding to two decimals embedded 
Percentage=[round((Genders_Count[0]/Total_Players)*100,2), round((Genders_Count[1]/Total_Players)*100,2), round((Genders_Count[2]/Total_Players)*100,2)]

#Create DataFrame
Gender_Demographics=pd.DataFrame({"Total Count":Genders_Count, "Percentage of Players": Percentage})
#Set index for Genders
Gender_Demographics.index=(["Male", "Female", "Other / Non-Disclosed"])

#print Dataframe 
Gender_Demographics
```

```python
#PURCHASING ANALYSIS-GENDER
#Basic Calculation for Purchase Count Per Gender
##M
Males_Purchase=df.loc[df["Gender"] =="Male","Purchase ID"].count()
##F
Females_Purchase=df.loc[df["Gender"] =="Female","Purchase ID"].count()
##O
Others_Purchase=df.loc[df["Gender"] =="Other / Non-Disclosed","Purchase ID"].count()

#Basic Calculation for Average Purchase Price Per Gender
##M
Males_Avg_Purchase=df.loc[df["Gender"] =="Male","Price"].mean()
##F
Females_Avg_Purchase=df.loc[df["Gender"] =="Female","Price"].mean()
##O
Others_Avg_Purchase=df.loc[df["Gender"] =="Other / Non-Disclosed","Price"].mean()

#Basic Calculation for Total Purchase Value Per Gender
##M
Males_Total_Purchase=df.loc[df["Gender"] =="Male","Price"].sum()
##F
Females_Total_Purchase=df.loc[df["Gender"] =="Female","Price"].sum()
##O
Others_Total_Purchase=df.loc[df["Gender"] =="Other / Non-Disclosed","Price"].sum()

#Basic Calculation for Avg Total Purchase per Person Per Gender
Per_Person =df.groupby(["Gender","SN"], as_index=False)['Price'].sum()
##M
Male_Avg_PPerson=Per_Person[Per_Person["Gender"]=="Male"]["Price"].mean()
##F
Female_Avg_PPerson=Per_Person[Per_Person["Gender"]=="Female"]["Price"].mean()
##O
Others_Avg_PPerson=Per_Person[Per_Person["Gender"]=="Other / Non-Disclosed"]["Price"].mean()

#Sumary Table for all data based on Genders
Purchase_Gender= pd.DataFrame({"Gender": ["Male","Female", "Other / Non-Disclosed"], "Purchase Count" : [Males_Purchase, Females_Purchase, Others_Purchase], 
                               "Average Purchase Price" : [Males_Avg_Purchase, Females_Avg_Purchase, Others_Avg_Purchase], "Total Purchase Value" : [Males_Total_Purchase, Females_Total_Purchase, Others_Total_Purchase],
                              "Average Total Purchase Per Person" : [Male_Avg_PPerson, Female_Avg_PPerson, Others_Avg_PPerson]})

#Cleaner Formatting - Dollar signs and comma at thousands placeholder
Purchase_Gender["Average Purchase Price"] = Purchase_Gender["Average Purchase Price"].map("${:.2f}".format)
Purchase_Gender["Total Purchase Value"] = Purchase_Gender["Total Purchase Value"].map("${:,}".format)
Purchase_Gender["Average Total Purchase Per Person"] = Purchase_Gender["Average Total Purchase Per Person"].map("${:.2f}".format)
#Display the Data Frame
Purchase_Gender
```

```python
#AGE DEMOGRAPHICS
#Create Bins
age_bins = [0, 10, 14, 19, 24, 29, 34, 39, 99]
#Create Group names for bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]

#Slice the date and add Age ranges to table
df["Age Ranges"] = pd.cut(df["Age"], age_bins, labels=group_names)

#Counts for bins
age_demos_totals = df["Age Ranges"].value_counts()

#Percent Calculations
age_demo_percents = age_demos_totals / Total_Players * 100

#Create DataFrame and print 
age_demos = pd.DataFrame({"Total Count": age_demos_totals, "Percent of Players": age_demo_percents})
age_demos = age_demos.sort_index()
age_demos
```

```python
#PURCHASE ANALYSIS-AGE
#use bins and cut data from above 

df["Age Ranges"] = pd.cut(df["Age"], age_bins, labels=group_names)


#Run Basic Calculation for Purchase Count
Purchase_Counts = df.groupby(["Age Ranges"]).count()["Price"]

#Run Basic Calculation for Average Purchase Price 
Avg_Purchase = df.groupby(["Age Ranges"]).mean()["Price"]

#Run Basic Calculation for Total Purchase Value
Total_Purchase_Value = df.groupby(["Age Ranges"]).sum()["Price"]

#Run Basic Calculation for Avg Total per person sort by unique SN to remove duplicates
Avg_Purchase_Pperson = Total_Purchase_Value / df.groupby(["Age Ranges"]).nunique()["SN"]

#Create DataFrame for Analysis and sort the index and print
Age_Per_Person_Analysis = pd.DataFrame({"Purchase Count": Purchase_Counts, "Average Purchase Price": Avg_Purchase, "Total Purchase Value": Total_Purchase_Value, "Avg Total Purchase per Person": Avg_Purchase_Pperson})

Age_Per_Person_Analysis = Age_Per_Person_Analysis.sort_index()
Age_Per_Person_Analysis
```

```python
#TOP SPENDERS
#group data by the SN
Top_Spenders=df.groupby("SN")

#count the totals of purcases by SN
Spender = Top_Spenders["Purchase ID"].count()

#calculate avg pruchase by SN
Avg_spender=Top_Spenders["Price"].mean()

#Calcualte Total Purchase Value for each SN
Total_Purchase=Top_Spenders["Price"].sum()

#make DataFrame for the above information with all calulations
Total_Top_Spenders = pd.DataFrame({"Purchase Count": Spender, "Average Purchase Price" : Avg_spender, "Total Purchase Value" : Total_Purchase})

#sort dataframe in descending order
sorted_spenders= Total_Top_Spenders.sort_values(["Total Purchase Value"], ascending=False)
sorted_spenders.head()
```

```python
#MOST POPULAR ITEMS
#create list with new dataframe for the items needed
Items=df[["Item ID", "Item Name", "Price"]]

#group by Item ID and Item Name
grouped=Items.groupby(["Item ID", "Item Name"])
#count the times purchased
purchased=grouped["Price"].count()

#find the sum value per item
purchased_value=(grouped["Price"].sum())

#calucalte the individual item price
item_cost= purchased_value/purchased
#create data frame for Most Popular Items
Most_Pop_Items= pd.DataFrame({ "Purchase Count": purchased , "Item Price" : item_cost , "Total Purchase Value" : purchased_value})
#Sort in decending order
popular_sort= Most_Pop_Items.sort_values(["Purchase Count"], ascending=False)
popular_sort.head()
```
